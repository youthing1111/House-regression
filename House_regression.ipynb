We drop column 'Fence', 'PoolQC' and 'MiscFeature' as it has 80% of data are null. 

```python
df_train = df_train.drop(columns=['Fence','PoolQC','MiscFeature'])
null_column = df_train.columns[df_train.isnull().sum()>0]
df_train[null_column]
```

```python
#Divide into categorical and numerical data
numerical_col = ['LotFrontage','MasVnrArea','GarageYrBlt']
categorical_col = [item for item in null_column if item not in numerical_col]
```

```python
#Replace missing categorical data as Unknown since there are just a few of it. 
df_train[categorical_col]  = df_train[categorical_col].apply(lambda x: x.fillna('Unknown'))
```

```python
cat_col = df_train.select_dtypes(include='object').columns
for var in cat_col:
    cat_list='var'+'_'+var
    cat_list = pd.get_dummies(df_train[var], prefix=var)
    cat_list = cat_list.astype(str)
    df1=df_train.join(cat_list)
    df_train=df1
data_vars=df_train.columns.values.tolist()
to_keep=[i for i in data_vars if i not in cat_col]
df_train_encoded=df_train[to_keep]
df_train_encoded.columns.values
```

```python
#Using regression and maximizing log-likelihood to fill missing data
null_data = df_train[df_train[['LotFrontage','MasVnrArea','GarageYrBlt']].isnull().any(axis=1)]
nonnull_data = pd.concat([null_data,df_train]).drop_duplicates(keep=False)
```

```python
for x in null_data.columns:
    if null_data[x].isnull().sum() > 0:
        print(x)
```

```python
null_data['MasVnrArea'].fillna(null_data['MasVnrArea'].mean(),inplace=True)
null_data['GarageYrBlt'].fillna(null_data['GarageYrBlt'].mean(),inplace=True)
```

```python
X =nonnull_data[[item for item in nonnull_data.columns if item not in ['LotFrontage','MasVnrArea','GarageYrBlt']]]
y1 = nonnull_data['LotFrontage']
```

```python
#Normalize numerical variables
from sklearn.preprocessing import StandardScaler
num_col = X.select_dtypes(include=np.number).columns
scaler = StandardScaler()
X[num_col] = scaler.fit_transform(X[num_col])
null_data[num_col] = scaler.transform(null_data[num_col])
```

```python
X['SalePrice']
```

```python
X.drop(columns=cat_col,inplace=True)
```

```python
#Lasso regression
from sklearn.linear_model import Lasso
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import scipy.stats as stats
def MLE_Norm(parameters):
    alpha,sd = parameters[0],parameters[1]
    X_train,X_test,y_train,y_test=train_test_split(X,y1, test_size=0.2, random_state=31)
    lasso = Lasso(alpha=alpha)
    lasso.fit(X_train,y_train)
    pred = lasso.predict(X_test)
    negLL = -np.sum(stats.norm.logpdf(y_test, loc=pred, scale=sd))
    return(negLL)
```

```python
#MLE
from scipy.optimize import minimize
guess = [0.1,2]
result = minimize(MLE_Norm, guess, method = 'Nelder-Mead', options={'disp': True})
result
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(X_train, y_train)
from sklearn.metrics import r2_score
y_pred = regressor.predict(X_test)
print(r2_score(y_test, y_pred))
print(np.mean(np.abs(y_test - y_pred)))
```
